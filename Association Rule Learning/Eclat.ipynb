```
data = pd.read_csv("Market_Basket_Optimisation.csv", header= None)
data.head()
```

```
temp = []
for i in range(len(data)):
  temp.append([str(data.values[i,j]) for j in range(0, 20)])
```

```
from apyori import apriori
ap = apriori(transactions = temp, min_support = 0.003, min_confidence = 0.2 , min_lift = 3, min_length = 2, max_length = 2)

results = list(ap)
results


```

```
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]

    return list(zip(lhs, rhs, supports))
final_pred = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support'])
final_pred
```

```
final_pred.nlargest(n = 10 , columns= "Support")
```
